# Percentile

```python
# NumPy is a Python library used for working with arrays.
# It also has functions for working in domain of linear algebra, fourier transform, and matrices.
#NumPy was created in 2005 by Travis Oliphant. It is an open source project and you can use it freely.
#NumPy stands for Numerical Python.
# For More details: https://www.w3schools.com/python/numpy/numpy_intro.asp
import numpy as np
```

```python
#First we check the single column for percentile of 25%, 50%, 75%, 99%, 100% 
print(np.percentile(dataset["ssc_p"],25))
print(np.percentile(dataset["ssc_p"],50))
print(np.percentile(dataset["ssc_p"],75))
print(np.percentile(dataset["ssc_p"],99))
print(np.percentile(dataset["ssc_p"],100))
```

```python
# when we come to for loop we haven't use np.percentile(dataset[columnname],99) directly 
# why because Null and NAN isn't handle 
# so we go for dataset.describe()[columnName]["50%"]. Here they already handled with Null and NAN 
# which mean they removed the null and NAN datas in dataset
# Index -> Title of the data performed. Quan(Numerical) -> Column of the data in dataset.  
# Added IQR title in the index next to Q4:100%

descrip=pd.DataFrame(index=["Mean","Median","Mode","Q1:25%","Q2:50%","Q3:75%","99%","Q4:100%","IQR","1.5rule","Lesser","Greater","Min","Max"], columns=quan)
for columnName in quan:
    descrip[columnName]["Mean"]=dataset[columnName].mean()
    descrip[columnName]["Median"]=dataset[columnName].median()
    descrip[columnName]["Mode"]=dataset[columnName].mode()[0]
    descrip[columnName]["Q1:25%"]=dataset.describe()[columnName]["25%"]
    descrip[columnName]["Q2:50%"]=dataset.describe()[columnName]["50%"]
    descrip[columnName]["Q3:75%"]=dataset.describe()[columnName]["75%"]
    descrip[columnName]["99%"]=np.percentile(dataset[columnName],99)
    descrip[columnName]["Q4:100%"]=dataset.describe()[columnName]["max"]
    descrip[columnName]["IQR"]=descrip[columnName]["Q3:75%"]-descrip[columnName]["Q1:25%"]
    descrip[columnName]["1.5rule"]=1.5*descrip[columnName]["IQR"]
    descrip[columnName]["Lesser"]=descrip[columnName]["Q1:25%"]-descrip[columnName]["1.5rule"]
    descrip[columnName]["Greater"]=descrip[columnName]["Q3:75%"]+descrip[columnName]["1.5rule"]
    descrip[columnName]["Min"]=dataset[columnName].min()
    descrip[columnName]["Max"]=dataset[columnName].max()
    
```

```python
# Frame the frequency table with required columns as per our needs
freqTable = pd.DataFrame(columns=["Unique_Value","Frequency","Relative_Frequency","Cumsum"])
freqTable
```

```python
# using the ssc_p column in dataset to get unique value, frequency, relative frequency and cumsum
freqTable["Unique_Value"]=dataset["ssc_p"].value_counts().index
freqTable["Frequency"]=dataset["ssc_p"].value_counts().values
freqTable["Relative_Frequency"]=(freqTable["Frequency"]/103)
freqTable["Cumsum"]=freqTable["Relative_Frequency"].cumsum()
```

```python
# see the dataframe of freqTable
freqTable
```

```python
# Make the frequency check in the function by passing 2 parameter as columnname and dataset 
# and return the FreqTable as result set
def freqTable(columnName,dataset):
    freqTable = pd.DataFrame(columns=["Unique_Value","Frequency","Relative_Frequency","Cumsum"])
    freqTable["Unique_Value"]=dataset[columnName].value_counts().index
    freqTable["Frequency"]=dataset[columnName].value_counts().values
    freqTable["Relative_Frequency"]=(freqTable["Frequency"]/103)
    freqTable["Cumsum"]=freqTable["Relative_Frequency"].cumsum()
    return freqTable
```

```python
# By passing columnName and dateset as parameter or input variables to freqTable function and check 
freqTable("ssc_p",dataset)
```

```python
# Make the descriptive check in the function by passing 2 parameter as dataset and quan 
# and return the descriptive as result set

def Univariate(dataset,quan):
    descrip=pd.DataFrame(index=["Mean","Median","Mode","Q1:25%","Q2:50%","Q3:75%","99%","Q4:100%","IQR","1.5rule","Lesser","Greater","Min","Max"], columns=quan)
    for columnName in quan:
        descrip[columnName]["Mean"]=dataset[columnName].mean()
        descrip[columnName]["Median"]=dataset[columnName].median()
        descrip[columnName]["Mode"]=dataset[columnName].mode()[0]
        descrip[columnName]["Q1:25%"]=dataset.describe()[columnName]["25%"]
        descrip[columnName]["Q2:50%"]=dataset.describe()[columnName]["50%"]
        descrip[columnName]["Q3:75%"]=dataset.describe()[columnName]["75%"]
        descrip[columnName]["99%"]=np.percentile(dataset[columnName],99)
        descrip[columnName]["Q4:100%"]=dataset.describe()[columnName]["max"]
        descrip[columnName]["IQR"]=descrip[columnName]["Q3:75%"]-descrip[columnName]["Q1:25%"]
        descrip[columnName]["1.5rule"]=1.5*descrip[columnName]["IQR"]
        descrip[columnName]["Lesser"]=descrip[columnName]["Q1:25%"]-descrip[columnName]["1.5rule"]
        descrip[columnName]["Greater"]=descrip[columnName]["Q3:75%"]+descrip[columnName]["1.5rule"]
        descrip[columnName]["Min"]=dataset[columnName].min()
        descrip[columnName]["Max"]=dataset[columnName].max()
    return descrip    

```

```python
# By passing needed parameter or input variables to Univariate function and check 
Univariate(dataset,quan)
```

```python
# Make the OutlierColumnName check in the function by passing 1 parameter as quan 
# And return the 2 result set of lesser and greater
# In Python, we can return more than one result set. 
# In this function we checked the 2 condition and stored in 2 different list so returning 2 result set

def OutlierColumnName(quan):
    for columnName in quan:
        if(descrip[columnName]["Min"]<descrip[columnName]["Lesser"]):
            lesser.append(columnName)
        if(descrip[columnName]["Max"]>descrip[columnName]["Greater"]):
            greater.append(columnName)  
    return lesser,greater        
```

```python
#By passing quan as parameter or input variables to OutlierColumnName function and check 
OutlierColumnName(quan)
```

```python
# Make the ReplacingOutlier check in the function by passing 2 parameter as lesser and greater 
# and return the dataset as result set

def ReplacingOutlier(lesser,greater):
    for columnName in lesser:
        dataset[columnName][dataset[columnName]<descrip[columnName]["Lesser"]]=descrip[columnName]["Lesser"]
    for columnName in greater:
        dataset[columnName][dataset[columnName]>descrip[columnName]["Greater"]]=descrip[columnName]["Greater"]
    return dataset   
```

```python
# Make the descriptive check in the function by passing 2 parameter as dataset and quan 
# and return the descriptive as result set

def Univariate(dataset,quan):
    descrip=pd.DataFrame(index=["Mean","Median","Mode","Q1:25%","Q2:50%","Q3:75%","99%","Q4:100%","IQR","1.5rule",
                                "Lesser","Greater","Min","Max","Kurtosis","Skew"], columns=quan)
    for columnName in quan:
        descrip[columnName]["Mean"]=dataset[columnName].mean()
        descrip[columnName]["Median"]=dataset[columnName].median()
        descrip[columnName]["Mode"]=dataset[columnName].mode()[0]
        descrip[columnName]["Q1:25%"]=dataset.describe()[columnName]["25%"]
        descrip[columnName]["Q2:50%"]=dataset.describe()[columnName]["50%"]
        descrip[columnName]["Q3:75%"]=dataset.describe()[columnName]["75%"]
        descrip[columnName]["99%"]=np.percentile(dataset[columnName],99)
        descrip[columnName]["Q4:100%"]=dataset.describe()[columnName]["max"]
        descrip[columnName]["IQR"]=descrip[columnName]["Q3:75%"]-descrip[columnName]["Q1:25%"]
        descrip[columnName]["1.5rule"]=1.5*descrip[columnName]["IQR"]
        descrip[columnName]["Lesser"]=descrip[columnName]["Q1:25%"]-descrip[columnName]["1.5rule"]
        descrip[columnName]["Greater"]=descrip[columnName]["Q3:75%"]+descrip[columnName]["1.5rule"]
        descrip[columnName]["Min"]=dataset[columnName].min()
        descrip[columnName]["Max"]=dataset[columnName].max()
        descrip[columnName]["Kurtosis"]=dataset[columnName].kurtosis()
        descrip[columnName]["Skew"]=dataset[columnName].skew()
    return descrip    
```
