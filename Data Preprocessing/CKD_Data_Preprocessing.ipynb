```python
import pandas as pd
```

```python
dataset = pd.read_csv("CKD.csv")
```

```python
dataset
```

```python
dataset.isnull().sum()
```

```python
def QualQuan(dataset):
    quan=[]
    qual=[]
    for ColumnName in dataset.columns:    
        if(dataset[ColumnName].dtype=='O'):
            qual.append(ColumnName)
        else:
            quan.append(ColumnName)
    return quan,qual #we can return more than one variable based on our result
```

```python
quan,qual=QualQuan(dataset)
```

```python
dataset[quan]
```

```python
dataset[qual]
```

```python
import numpy as np
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy='mean')
imp.fit(dataset[quan])
Quan_dataset=imp.transform(dataset[quan])
```

```python
Quan_dataset
```

```python
Quan_dataset=pd.DataFrame(Quan_dataset,columns=quan)
```

```python
Quan_dataset
```

```python
dataset[qual]
```

```python
import numpy as np
from sklearn.impute import SimpleImputer
impu = SimpleImputer(missing_values=np.nan, strategy='mean')
impu.fit(dataset[qual])
Qual_dataset=impu.transform(dataset[qual])
```

```python
Qual_dataset
```

```python
Qual_dataset=pd.DataFrame(Qual_dataset,columns=qual)
```

```python
Qual_dataset
```
